# Data Preprocessing

## Reformat Gesture Data

```python
def convert_to_array(gesture_df):
    
    def flatten_dict_col(df, column):
        return df[column].apply(pd.Series)

    acceleration_df = flatten_dict_col(gesture_df, 'acceleration')
    acceleration_df = acceleration_df[['x', 'y', 'z']]
    
    rotation_rate_df = flatten_dict_col(gesture_df, 'rotationRate')
    rotation_rate_df = rotation_rate_df[['x', 'y', 'z']]
    
    combined_df = pd.concat([acceleration_df, rotation_rate_df], axis=1)
    
    gesture_array = combined_df.to_numpy(dtype=np.double)
    
    return gesture_array
```

```python
storedGesture_array = convert_to_array(storedGesture_df)
newGesture_array = convert_to_array(newGesture_df)
```

```python
print(f"Stored Gesture Shape: {storedGesture_array.shape}")
print(f"New Gesture Shape: {newGesture_array.shape}")
```

## Sync Gesture Data

We need to make the stored gesture data and new gesture data have the same number of data points.

### Method 1: Cut the gesture data to min data points

```python
min_data_points = min(len(storedGesture_array), len(newGesture_array))

if (len(storedGesture_array) > len(newGesture_array)):
    storedGesture_array = storedGesture_array[:min_data_points]
else:
    newGesture_array = newGesture_array[:min_data_points]
```

```python
print(f"Stored Gesture Shape: {storedGesture_array.shape}")
print(f"New Gesture Shape: {newGesture_array.shape}")
```

### Method 2: Add dummy data points to min data points

```python
if (len(storedGesture_array) > len(newGesture_array)):
    for i in range(len(storedGesture_array) - len(newGesture_array)):
        newGesture_array = np.append(newGesture_array, [newGesture_array[-1]], axis=0)
else:
    for i in range(len(newGesture_array) - len(storedGesture_array)):
        storedGesture_array = np.append(storedGesture_array, [storedGesture_array[-1]], axis=0)
```

```python
print(f"Stored Gesture Shape: {storedGesture_array.shape}")
print(f"New Gesture Shape: {newGesture_array.shape}")
```

# Gesture Matching

## Method 1: DTW

```python
from dtaidistance import dtw_ndim
from dtaidistance import dtw
from dtaidistance import dtw_visualisation as dtwvis
```

```python
%%time
d = dtw_ndim.distance(storedGesture_array, newGesture_array)
d
```

```python
print(f"Stored Gesture Shape: {storedGesture_array.shape}")
print(f"New Gesture Shape: {newGesture_array.shape}")
```

```python
from matplotlib.gridspec import GridSpec, GridSpecFromSubplotSpec

n_dimensions = 6
dimension_labels = ['Acceleration X', 'Acceleration Y', 'Acceleration Z', 'Rotation X', 'Rotation Y', 'Rotation Z']

fig = plt.figure(figsize=(15, 10))

outer_grid = GridSpec(2, 3, wspace=0.2, hspace=0.3)

for i in range(n_dimensions):
    inner_grid = GridSpecFromSubplotSpec(2, 1, subplot_spec=outer_grid[i], hspace=0.4)
    
    path = dtw_ndim.warping_path(storedGesture_array[:, i], newGesture_array[:, i])

    ax1 = plt.Subplot(fig, inner_grid[0])
    fig.add_subplot(ax1)

    ax2 = plt.Subplot(fig, inner_grid[1])
    fig.add_subplot(ax2)
    
    dtwvis.plot_warping(storedGesture_array[:, i], newGesture_array[:, i], path, fig=fig, axs=[ax1, ax2])

    ax1.set_title(dimension_labels[i])

plt.show()
```
